```python
xl = pd.ExcelFile('data/2015-16/Annex 18 - Reconciliation sheets Gems & Jade 15-16.xlsx')
sheets_list = xl.sheet_names
#df = xls.parse('Sheet1', skiprows=4, index_col=None, na_values=['NA'])
```

```python
def join_column_titles(text):
    #print(text)
    to_return = ""
    for t in text:
        #print(t)
        if not isinstance(t, float):
            to_return = '.'.join([to_return,(''.join(i for i in t if ord(i)<128))])
    #print("FINAL: " + to_return)
    
    # remove first '.' from the title
    return to_return[1:]
```

```python
df_all = pd.DataFrame({'company_name' : [],'name_of_revenue_stream' : [],'value_reported' : []})


for s in sheets_list:
    try:
        #df = xl.parse('C (1)', skiprows=4, index_col=None, na_values=['NA'])
        name_df = xl.parse(s, parse_cols = 'C:E')
        company_name = name_df.columns[2]
        company_number = name_df[name_df['Company name:'] == 'Registry number'][company_name][0]
        print(company_name, company_number)

        df = xl.parse(s, skiprows=4, skip_footer=37, parse_cols = 'B:M', header=None)
        df.iloc[0:2] = df.iloc[0:2].fillna(method='ffill', axis=1)
        df.columns = df.iloc[0:2].apply(join_column_titles, axis=0)
        df = df.iloc[2:33]

        name = map(lambda x: ''.join(e for e in x if e.isalnum()) , company_name.split(' '))
        name = ' '.join(w for w in name)
        name = re.sub( '\s+', ' ', name.strip())
        filename = re.sub( '\s+', '_', name.strip())+'.csv'
    #         df.to_csv(filename)

        print(filename)
        
        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Sale Split', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[9,'Per Government.Final'][0]]})])
        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Sales Royalties', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[10,'Per Government.Final'][0]]})])
        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Emporium Fees', \
                          'paid_to' : "Central Comittee", \
                          'value_reported' : [df.loc[25,'Per Government.Final'][0]]})])

        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Other significant payments', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[23,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Production Royalties', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[7,'Per Government.Final'][0]]})])

        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Production Royalties', \
                          'paid_to' : "Department of Mines", \
                          'value_reported' : [df.loc[8,'Per Government.Final'][0]]})])

        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Permit Fees', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[12,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Incentive Fees', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[13,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Commercial Tax', \
                          'paid_to' : "Myanmar Gems Enterprise", \
                          'value_reported' : [df.loc[14,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Corporate Income Tax', \
                          'paid_to' : "Internal Revenue Department", \
                          'value_reported' : [df.loc[19,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Commercial Tax', \
                          'paid_to' : "Internal Revenue Department", \
                          'value_reported' : [df.loc[20,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Customs Duties', \
                          'paid_to' : "Customs Department", \
                          'value_reported' : [df.loc[21,'Per Government.Final'][0]]})])


        df_all = pd.concat([df_all, \
            pd.DataFrame({'company_name' : [name], \
                          'name_of_revenue_stream' : 'Sales Royalties', \
                          'paid_to' : "Internal Revenue Department", \
                          'value_reported' : [df.loc[22,'Per Government.Final'][0]]})])
        #         print(df.columns)
    except:
        print("error parsing names from sheet \"" + s + "\"")
```
